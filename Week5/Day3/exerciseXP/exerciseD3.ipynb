Install Matplotlib and import it into a Python script using import matplotlib.pyplot as plt.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Create a basic line plot using the following data, which represents the demand for a product at different price points.

```python
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand

figure1, axes1 = plt.subplots()
axes1.plot(x, y)
```

Create a basic scatter plot using the following data, which represents children’s heights at different ages:


```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

figure2, axes2 = plt.subplots()
axes2.scatter(x, y)
```

Modify the line plot from Exercise 2 to include a title (“Product Demand vs Price”), and labels for the x-axis (“Price”) and y-axis (“Demand”).

```python
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand

figure3, axes3 = plt.subplots()
axes3.set(title="Product Demand vs Price", xlabel="Price", ylabel="Demand")
axes3.plot(x, y)
```

Modify the scatter plot from Exercise 3 to include a title (“Children’s Height vs Age”), and labels for the x-axis (“Age (years)”) and y-axis (“Height (cm)”).

```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

figure4, axes4 = plt.subplots()
axes4.set(title="Children’s Height vs Age", xlabel="Age (years)", ylabel="Height (cm)")
axes4.scatter(x, y)
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

Create a line plot and scatter plot representing your favorite mathematical functions. Change the line color to tab:blue for the line plot and use a marker of your choice for the scatter plot.

```python
fig, axs = plt.subplots()
axs.plot(x, y1, color='tab:blue')
plt.show()
fig, axs = plt.subplots()
axs.scatter(x, y2, marker='*')
plt.show()
```

Plot the line plot and scatter plot from Exercise 7 on the same figure using subplots. Use plt.subplots to create a figure with two subplots.

```python
fig, axs = plt.subplots(2, 1)

axs[0].plot(x, y1, color='tab:blue')
axs[1].scatter(x, y2, marker='*')
plt.show()
```

Modify the plots from Exercise 8 to include a legend. Use the legend function from Matplotlib.

```python
fig, axs = plt.subplots(2, 1)

axs[0].plot(x, y1, color='tab:blue', label='y=sin(x)')
axs[0].legend(loc='upper left', shadow=True)
axs[1].scatter(x, y2, marker='*', label='y=cos(x)')
axs[1].legend(loc='center right', shadow=True)
plt.show()
```
