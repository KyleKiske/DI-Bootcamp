## Task 1

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 20, 100)
y = x ** 2

figure, axes = plt.subplots()
axes.set(title="Plot of x vs y (y=x^2)", xlabel="x", ylabel="y")
axes.set_ylim(0,400)
axes.set_xlim(0,20)
axes.plot(x, y)
plt.show()

figure.savefig('dcD3_1.png')
```

## Task 2

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

```python
chipo
```

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x:  float(x.replace('$','')))
chipo
```

```python
x = chipo['item_price']
y = chipo['quantity']

figure, axes = plt.subplots()
axes.set(title="Relationship between Item Price and Quantity Ordered", xlabel="Item Price", ylabel="Quantity Ordered")
axes.scatter(x, y)
plt.show()

figure.savefig('dcD3_2.png')
```

## Task 3

```python
chipo = chipo[['item_name','quantity']].groupby('item_name').sum('quantity')
```

```python
for_graph = chipo.sort_values(by=['quantity'], ascending=False).head(5)
for_graph
```

```python
figure, axes = plt.subplots()
axes.set(xlabel='Item Name', ylabel='Total Quantity Ordered')
axes.bar( for_graph.index, for_graph['quantity'])
plt.xticks(rotation=45)
plt.show()

figure.savefig('dcD3_3.png')
```
