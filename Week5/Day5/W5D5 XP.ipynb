```python
normalized_df=(boston-boston.min())/(boston.max()-boston.min())
normalized_df
```

#### Categorical Variables: The Boston Housing dataset doesnâ€™t contain categorical variables (variables with set categories), but itâ€™s useful to know how to handle them. Practice this by creating a fake categorical variable.

```python
normalized_df['age_category'] = pd.cut(normalized_df['age'],bins=[0.0,0.30,0.60,0.90],labels=['new','old','very_old'])
normalized_df.head()
```

#### Correlations: See how much each variable affects the house price (MEDV). This is done by calculating the correlation matrix (a table showing correlation coefficients between variables).

```python
correlation_matrix = normalized_df.corr(numeric_only=True)
correlation_matrix['medv'].sort_values()
```

```python
sns.heatmap(correlation_matrix[['medv']].sort_values(by='medv'), vmin=-1, vmax=1, annot=True)
```

## Part II

#### Analyse the relationship between Property Age and Price

```python
new_properties = boston[boston['age'] <= 35]
old_properties = boston[(boston['age'] <= 70) & (boston['age'] > 35)]
very_old_properties = boston[boston['age'] > 70]

fig, axs = plt.subplots(3, 1)
fig.tight_layout()
axs[0].set(xlabel='median value', ylabel='age')
axs[0].scatter(new_properties['medv'], new_properties['age'])
axs[1].set(xlabel='median value', ylabel='age')
axs[1].scatter(old_properties['medv'], old_properties['age'])
axs[2].set(xlabel='median value', ylabel='age')
axs[2].scatter(very_old_properties['medv'], very_old_properties['age'])
plt.show()
```

## Part III

#### Instruction : Analyse the Neighborhood Crime Rate Impact on Prices

```python
boston['CRIM_BIN'] = pd.cut(boston['crim'], bins=[0, 1, 5, 10, 20, max(boston['crim'])], labels=['0-1', '1-5', '5-10', '10-20', '20+'])
boston
```

```python
avg_MEDV_per_CRIM_BIN = boston.groupby('CRIM_BIN')['medv'].mean()
avg_MEDV_per_CRIM_BIN
```

```python
fig, axs = plt.subplots()
avg_MEDV_per_CRIM_BIN.plot(kind='line', ax=axs, xlabel='average crime', ylabel='median value')
plt.show()
```
