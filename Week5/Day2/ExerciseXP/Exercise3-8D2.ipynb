# ðŸŒŸ Exercise 4: Merge (All_data, All_data_col, Data3)

##### Create three DataFrames based on the given raw data (define what the raw data looks like and what data should be included in â€˜data1â€™, â€˜data2â€™, and â€˜data3â€™).

```python
data1 = pd.DataFrame({
    'Animal': ['Falcon', 'Falcon', 'Parrot', 'Parrot'],
    'Max Speed': [380., 370., 24., 26.],
    'Weight': [1, 1.5, 0.3, 0.5]
})
data2 = pd.DataFrame({
    'Animal': ['Bear', 'Deer', 'Tiger', 'Honey badger'],
    'Max Speed': [40., 65., 50., 20.],
    'Weight': [350, 160, 150, 8]
})
data3 = pd.DataFrame({
    'Animal': ['Wolf', 'Fox', 'Boar'],
    'Max Speed': [65., 50., 40.],
    'Weight': [45, 7, 80]
})
```

Merge the two dataframes â€˜data1â€™ and â€˜data2â€™ along rows using the concat() function and assign it to â€˜all_dataâ€™.

```python
all_data = pd.concat([data1, data2])
```

```python
all_data
```

Merge â€˜data1â€™ and â€˜data2â€™ along columns using concat() and assign it to â€˜all_data_colâ€™.

```python
all_data_col = pd.concat([data1, data2], axis=1)
```

```python
all_data_col
```

## ðŸŒŸ Exercise 6: Creating Series And DataFrames (Pokemon)

```python
pokemon_dict = {'evolution': ['Ivysaur', 'Charmeleon', 'Wartortle', 'Metapod'],
    'hp': [45, 39, 44, 45],
    'name': ['Bulbasaur', 'Charmander', 'Squirtle', 'Caterpie'],
    'pokedex': ['yes', 'no', 'yes', 'no'],
    'type': ['grass', 'fire', 'water', 'bug']}
```

```python
pokemon = pd.DataFrame.from_dict(pokemon_dict)
```

```python
pokemon
```

#### Rearrange the columns in the order: â€˜nameâ€™, â€˜typeâ€™, â€˜hpâ€™, â€˜evolutionâ€™, â€˜pokedexâ€™ using simple DataFrame indexing.

```python
cols = pokemon.columns.tolist()
cols
```

```python
cols = [cols[2]] + [cols[4]] + [cols[1]] + [cols[0]] + [cols[3]]
cols
```

```python
pokemon[cols]
```
