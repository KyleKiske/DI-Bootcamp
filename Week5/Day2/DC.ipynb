Find out the average calorie content per drink category.

```python
result = starbucks[['Beverage_category','Calories']].groupby(['Beverage_category']).agg(np.mean)
result
```

Create a new column ‘caffeine_to_calories’ that represents the ratio of caffeine to calories in each drink.

```python
new_df = starbucks
new_df
```

```python
starbucks.drop(starbucks[(starbucks['Caffeine (mg)'] == 'Varies') | (starbucks['Caffeine (mg)'] == 'varies')].index, inplace = True)
```

```python
starbucks['caffeine_to_calories'] = starbucks['Caffeine (mg)'].astype(float) / starbucks['Calories']
```

```python
starbucks = starbucks.dropna()
```

Find the drink category that, on average, has the highest ‘caffeine_to_calories’ ratio.

```python
result = starbucks[['Beverage_category','caffeine_to_calories']].groupby(['Beverage_category']).agg(np.mean)
result = result.dropna()
result
```

Create a horizontal bar plot that shows the average ‘caffeine_to_calories’ ratio for each drink category. Use matplotlib’s barh() function for this.

```python
result['caffeine_to_calories'] = result['caffeine_to_calories'].astype(float)
```

```python
plt.barh(result.index, result['caffeine_to_calories'])
plt.show()
```
